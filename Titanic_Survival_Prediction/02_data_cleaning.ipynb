```python
import pandas as pd
from utils import cabin_embarked_transformer, title_group_creation, fit_age_imputation, transform_age_imputation, family_size_creation, merge_test
```

```python
df = pd.read_csv('data/train.csv')
df_merged = merge_test()
df_merged
```

```python
df = cabin_embarked_transformer(df)
df_merged = cabin_embarked_transformer(df_merged)
df
```

```python
df_cleaned = family_size_creation(df_cleaned)
df_merged_cleaned = family_size_creation(df_merged_cleaned)
df_merged_cleaned
```
